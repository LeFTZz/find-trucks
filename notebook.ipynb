```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
df = pd.read_csv('Mobile_Food_Facility_Permit.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df = df[df['Status']=='APPROVED']
df = df[(df['Latitude']!=0) | (df['Longitude']!=0)]
df.shape
```

```python
df['Latitude'].dtype, df['Longitude'].dtype
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['Longitude'], df['Latitude']))
gdf.head()
```

```python
input_lat = 37.5
input_lon = -122.5
gdf['dis_input'] = gdf['geometry'].distance(Point(input_lon, input_lat))
top5 = gdf.sort_values('dis_input', ascending=True).head(5)
top5
```

```python
import json
result = top5.to_json()
parsed = json.loads(result)
parsed
```
